##Movie table

```python
movies_metadata = pd.read_csv('./data/movies_metadata.csv', 
                              usecols=['belong_to_collection',
                                       'id',
                                       'genres', 
                                       'imdb_id', 
                                       'title',
                                       'overview', 
                                       'release_date'],
                              dtype={'id': str})
movies_metadata.head()
```

```python
# drop NA
movies_metadata.dropna(axis=0, how='any',inplace=True)
# drop duplicates
movies_metadata.drop_duplicates(inplace=True)
```

```python
Movie = pd.DataFrame(data=
                     {'movie_id': movies_metadata['id'].to_list(), 
                     'imdb_id': movies_metadata['imdb_id'].to_list(), 
                     'title': movies_metadata['title'].to_list(),
                     'release_date': movies_metadata['release_date'].to_list()})
Movie.head()
```

#User table

```python
ratings = pd.read_csv('/content/drive/MyDrive/DSS/project/data/database2/Ratings.csv', 
                      dtype={'movie_id': int, 'user_id':int}, nrows=1000000)
ratings.head()
```

```python
# drop na and duplicates
ratings.dropna(axis=0, how='any', inplace=True)
ratings.drop_duplicates(inplace=True)
```

```python
len(ratings['movie_id'].unique())
```

```python
max(ratings['user_id'])
```

```python
nonExistingMovies = set(ratings['movie_id']).difference(Movie['movie_id'])
len(nonExistingMovies)
```

```python
existingMovies = set(ratings['movieId']).difference(nonExistingMovies)
len(existingMovies)
```

```python
index_movieId = dict()
for i in ratings['movieId']:
    index_movieId[i] = []
```

```python
rating_indexes = ratings.index
for i in range(len(rating_indexes)):
    index_movieId[ratings['movieId'][i]].append(rating_indexes[i])
```

```python
dropIndexes = []
for i in list(nonExistingMovies):
    dropIndexes.extend(index_movieId[i])
dropIndexes[:10]
```

```python
print(len(ratings))
ratings.drop(index=dropIndexes, inplace=True)
print(len(ratings))
```

```python
ratings.head()
```

```python
userId = []
movieId = []
ratingList = []
timestamp = []
for i in list(existingMovies):
    print(i)
    coredf = ratings[ratings['movieId'] == i]
    userId += coredf['userId'].to_list()
    movieId += coredf['movieId'].to_list()
    ratingList += coredf['rating'].to_list()
    timestamp += coredf['timestamp'].to_list()  
```

```python
Ratings = pd.DataFrame(data={
                    'user_id': ratings['userId'],
                    'movie_id': ratings['movieId'],
                    'rating': ratings['rating'],
                    'timestamp': ratings['timestamp']})
Ratings.head()
```

```python
uniqueUserId = Ratings['user_id'].unique()
```

```python
User = pd.DataFrame(data={
    'user_id': uniqueUserId,
    'user_name': uniqueUserId,
    'password': uniqueUserId
})
User.head()
```

```python
len(User)
```

```python
Movie.to_csv('database2/Movies.csv', index=False)
User.to_csv('database2/Users.csv', index=False)
Ratings.to_csv('database2/Ratings.csv', index=False)
```

```python
accept_json_string_genres = []
for g in movies_metadata['genres']:
    accept_json_string_genres.append(g.replace("\'", "\""))
accept_json_string_genres[:3]
```

```python
json_genres_list = []
for genre_json_string in accept_json_string_genres:
    json_genres_list.append(json.loads(genre_json_string))
json_genres_list[:3]
```

```python
movie_id = Movie['movie_id'].to_list()
```

```python
genre_id_list = []
genre = {}
repeate_movie_id = []
for (i, json_genres) in enumerate(json_genres_list):
    for json_genre in json_genres:
        repeate_movie_id.append(movie_id[i])
        genre_id = json_genre['id']
        genre_id_list.append(genre_id)
        genre[genre_id] = json_genre['name']
```

```python
print(genre)
```

```python
MovieGenre = pd.DataFrame(data={
                'movie_id': repeate_movie_id,
                'genre_id': genre_id_list
                })
MovieGenre.head()
```

```python
len(MovieGenre['movie_id'].unique())
```

```python
len(Movie)
```

```python
Genre = pd.DataFrame(data={
    'genre_id': genre.keys(),
    'genre_name': genre.values()
})
Genre.head()
```

```python
MovieGenre.to_csv('database2/MovieGenres.csv', index=False)
Genre.to_csv('database2/Genres.csv', index=False)
```
